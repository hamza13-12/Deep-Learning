```python
import cv2
import numpy as np
from keras.models import load_model

model = load_model("best_model2.h5")
# Open webcam
cap = cv2.VideoCapture(0)

while True:
    # Read frame from the webcam
    ret, frame = cap.read()

    # Check if the frame is valid
    if not ret:
        print("Error: Couldn't read frame from the webcam.")
        break

    # Preprocess the frame (resize, normalize, expand dimensions)
    resized_frame = cv2.resize(frame, (220, 220))
    normalized_frame = resized_frame.astype('float32') / 255.0
    input_frame = np.expand_dims(normalized_frame, axis=0)

    # Get prediction from the model
    prediction = model.predict(input_frame)

    # Get the predicted class label
    predicted_class = np.argmax(prediction, axis=1)
    class_label = ['Hamza', 'Subhan', 'Unknown'][predicted_class[0]]

    # Put the predicted class label on the frame
    cv2.putText(frame, class_label, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)

    # Display the frame with prediction
    cv2.imshow('Webcam Prediction', frame)

    # Break the loop if 'q' or 'esc' key is pressed
    if cv2.waitKey(1) & 0xFF in (ord('q'), 27):
        break

# Release the webcam and close all windows
cap.release()
cv2.destroyAllWindows()
```
